This code uses the pd.read_csv() function from the pandas library to read a CSV file located at "/content/Frailty/raw_data.csv" and loads the data into a DataFrame named 'data'. Please replace the file path with the correct one if it's in a different location.

```python
import pandas as pd
```

```python
 data= pd.read_csv("/content/Student Performance/raw_data/StudentsPerformance.csv")
 print(data.to_string())
```

This code uses the isnull() method to identify missing values in the DataFrame 'data' and then sums the count of missing values for each column using sum(). Finally, it prints the count of missing values for each column to the console.

```python
data.isnull().sum()
```

This code calculates the average score for each student by averaging the 'math score', 'writing score', and 'reading score' columns and adds a new column 'Average_Score'. Then, it saves the DataFrame to a new CSV file at the specified path and displays the first few rows of the DataFrame using head(). Please replace the file path with the correct one if it's in a different location.

```python
 data['Average_Score'] = (data['math score'] + data['writing score'] + data['reading score'])/3
 data.to_csv("/content/Student Performance/clean_data/CleanData_Student_Performance.csv")
 data.head()
```

This code uses Seaborn and Matplotlib to create a countplot, visualizing the distribution of math scores based on race/ethnicity. The resulting plot is saved as an image file, and the plot is displayed using plt.show().

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
palette = "Set2"
sns.countplot(data=data, x='race/ethnicity', palette=palette)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Math Score')
plt.savefig('/content/Student Performance/results/bar chart.png')
plt.show()


```

This code uses Matplotlib to create a histogram with three categories: 'Race/Ethnicity', 'Gender', and 'Parental Education'. The resulting histogram is saved as an image file, and the plot is displayed using plt.show().

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen']
plt.hist([data['race/ethnicity'], data['gender'], data['parental level of education']], bins=20, color=colors, edgecolor='black', label=['Race/Ethnicity', 'Gender', 'Parental Education'])
plt.xlabel('All 3 types of categories')
plt.ylabel('Values')
plt.title('Histogram')
plt.legend()
plt.savefig('/content/Student Performance/results/histogram.png')
plt.show()

```

This code uses Seaborn and Matplotlib to create a boxplot, visualizing the distribution of math scores based on the test preparation course. The resulting boxplot is saved as an image file, and the plot is displayed using plt.show().

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
palette = {'completed': 'skyblue', 'none': 'lightcoral'}
sns.boxplot(x='test preparation course', y='math score', data=data, palette=palette)
plt.title('Math Score by Test Preparation Course')
plt.savefig('/content/Student Performance/results/boxplot.png')
plt.show()
```

This code uses Matplotlib to create a pie chart representing the distribution of math, reading, and writing scores. The resulting pie chart is saved as an image file, and the plot is displayed using plt.show().

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Assuming 'data' is your DataFrame
colors = ['red', 'blue', 'lightgreen']
labels = ['Math Score', 'Reading Score', 'Writing Score']
sizes = [data['math score'].count(), data['reading score'].count(), data['writing score'].count()]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Scores')
plt.savefig('/content/Student Performance/results/pie_chart.png')
plt.show()
```

This code uses Seaborn and Matplotlib to create a scatter plot, visualizing the relationship between reading scores and writing scores with gender represented by different colors. The resulting scatter plot is saved as an image file, and the plot is displayed using plt.show().

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

a = sns.scatterplot(data=data, x='reading score', y='writing score', hue='gender', palette='Set2', s=100, alpha=0.7)

plt.title('Scatter Plot of Reading Score vs Writing Score with Gender Hue')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')

plt.legend(title='Gender', loc='upper right')

plt.savefig('/content/Student Performance/results/scatter_plot.png')
plt.show()
```
